# Model Training

```python
## Independent And dependent features
x=df_copy.drop('FWI',axis=1)
y=df[['FWI']]
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
x_train.corr()
```

# Feature Selection

```python
# Check for multicollinearity
def correlation(dataset,threshold):
    '''This function returns that feature names which supports multicollinearility'''
    col_mulcol=set()
    corr_set=dataset.corr()
    for i in range(len(dataset.columns)):
        for j in range(i):
            if abs(corr_set.iloc[i,j])>threshold:
                colname=corr_set.columns[i]
                col_mulcol.add(colname)
    return col_mulcol
```

```python
## threshold--Domain expertise
mul_corr_features=correlation(x_train,0.85)
```

```python
mul_corr_features
```

```python
x_train.shape,x_test.shape
```

```python
## drop features when correlation is more than 0.85
x_train.drop(mul_corr_features,axis=1,inplace=True)
x_test.drop(mul_corr_features,axis=1,inplace=True)
x_train.shape,x_test.shape
```

## Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
fig=plt.figure(figsize=(10,5))
fig1=fig.add_subplot(2,2,1)
fig1.boxplot(x_train)
plt.title("x_train before standardization")
fig2=fig.add_subplot(2,2,2)
fig2.boxplot(x_train_scaled)
plt.title("x_train after standardization")
```

# Regression Models 

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
linear=LinearRegression()
linear.fit(x_train_scaled,y_train)
y_pred=linear.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(x_train_scaled,y_train)
y_pred=lasso.predict(x_test_scaled)

mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

## Ridge Regression model

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(x_train_scaled,y_train)
y_pred=ridge.predict(x_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

## Elasticnet Regression


```python
from sklearn.linear_model import ElasticNet
elastic=ElasticNet()
elastic.fit(x_train_scaled,y_train)
y_pred=elastic.predict(x_test_scaled)

mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```
